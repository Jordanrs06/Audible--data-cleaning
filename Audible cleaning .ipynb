```python
new_df = df.copy()
```

```python
import re
def rated_or_not(text):
    if text == 'Not rated yet':
        return np.nan
    else:
        rating = re.search(r"(\d+)(?=\s+ratings)", text)
        if rating:
            return (rating.group(1))
```

```python
new_df.insert(7,'Ratings',new_df['stars'].apply(rated_or_not))
```

```python
new_df
```

```python
def stars_out_of_5(text):
    if text == 'Not rated yet':
        return '0'
    else:
        stars = re.search(r"(\d+(\.\d+)?)\s+out\s+of\s+\d+\s+stars", text)
        if stars:
            return (stars.group(1))
```

```python
new_df['stars'] = new_df['stars'].apply(stars_out_of_5)
```

```python
new_df = new_df.rename(columns={'stars':'stars out of 5'})
```

```python
new_df['stars out of 5'] = new_df['stars out of 5'].astype('float')
```

```python
new_df
```

```python
new_df.info()
```

```python
new_df['author'] = new_df['author'].str[10:]
new_df['narrator'] = new_df['narrator'].str[11:]
```

```python
new_df['author'] = new_df['author'].str.strip(" ").str.strip(",").str.split(',')
new_df['author'] = new_df['author'].apply(lambda x:str(x).strip(" ").rstrip(']').lstrip('[')).apply(lambda x:re.sub("'","",x)).apply(lambda x:re.sub('"',"",x))
new_df['narrator'] = new_df['narrator'].str.strip(" ").str.strip(",").str.split(',')
new_df['narrator'] = new_df['narrator'].apply(lambda x:str(x).strip(" ").rstrip(']').lstrip('[')).apply(lambda x:re.sub("'","",x)).apply(lambda x:re.sub('"',"",x))
```

```python
new_df
```

```python
def formating_of_capital_letters(text):
    return re.sub(r"(?<=[a-z])(?=[A-Z])|,\s*", lambda m: " " if m.group() == "" else ", ", text)
```

```python
new_df['author'] = new_df['author'].apply(formating_of_capital_letters)
new_df['narrator'] = new_df['narrator'].apply(formating_of_capital_letters)
```

```python
new_df.sample(5)
```

```python
new_df['author'] = new_df['author'].str.title()
new_df['narrator'] = new_df['narrator'].str.title()
```

```python
new_df['author'] = new_df['author'].apply(lambda x:'Unknown' if x in ['Anonymous','Uncredited','Variousauthors'] else x)
new_df['narrator'] = new_df['narrator'].apply(lambda x:'Unknown' if x in ['Anonymous','Uncredited','Variousauthors'] else x)
```

```python
new_df[['author','narrator']].isin(['Anonymous','Uncredited','Variousauthors']).sum()
```

```python
time_pattern = re.compile(r"(?:(\d+)\s*hrs?)?\s*(?:and)?\s*(?:(\d+)\s*min)?")
def parse_time_string(time_str):
    match = time_pattern.search(time_str)
    if match:
        hours = int(match.group(1) or 0)
        minutes = int(match.group(2) or 0)
        return pd.Timedelta(hours=hours, minutes=minutes)
```

```python
new_df['time'] = new_df['time'].apply(parse_time_string)
```

```python
new_df['price'] = new_df['price'].str.replace(',',"").str.replace('Free','0')
```

```python
new_df['price'] = new_df['price'].astype('float')
```

```python
new_df['name'] = new_df['name'].str.replace('Book','Vol.').str.replace('#','Vol.')
```

```python
new_df['language'] = new_df['language'].str.capitalize()
```
